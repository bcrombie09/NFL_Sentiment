```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Data/initial_dataset.csv")
df.head()
```

```python
df['subreddit'].unique() # seeing all of the subreddits. It looks like data for the Broncos is missing. Will work around it
```

```python
# List of winning teams - 16
winning_teams = ['falcons','ravens','Browns','cowboys','detroitlions','GreenBayPackers','Jaguars','raiders','LosAngelesRams','miamidolphins','Saints','nyjets','eagles','49ers','buccaneers','Commanders']
print(len(winning_teams))
```

```python
# List of losing teams - 16
losing_teams = ['AZCardinals','buffalobills','panthers','ChicagoBearsNFL','bengals','DenverBroncos','Texans','Colts','KansasCityChiefs','Chargers','minnesotavikings','Patriots','NYGiants','steelers','Seahawks','Tennesseetitans']
print(len(losing_teams))
```

```python
# List of expected wins - according to fanduel
expected_win = ['falcons','ravens','buffalobills','ChicagoBearsNFL','bengals','cowboys','DenverBroncos','Jaguars','KansasCityChiefs','Chargers','minnesotavikings','Saints','eagles','49ers','Seahawks','Commanders']
print(len(expected_win))
```

```python
# List of expected loss - according to fanduel
expected_loss = ['AZCardinals','panthers','Browns','detroitlions','GreenBayPackers','Texans','Colts','raiders','LosAngelesRams','miamidolphins','NYGiants','Patriots','nyjets','steelers','buccaneers','Tennesseetitans']
print(len(expected_loss))
```

```python
# Adding actual result to dataframe
def assign_result(team):
    if team in winning_teams: # Of the team won, return 1. If the team lost, return 0
        return 1
    elif team in losing_teams:
        return 0
    else:
        return None  # For the DenverBroncos/nan team

df['result'] = df['subreddit'].apply(lambda x: assign_result(x)) # creating a new column 'result' for the expected result
```

```python
df.head()
```

```python
def assign_expected(team):  # Function for assigning the expected result from fanduel
    if team in expected_win:   # 1 for expected win, 0 for expected loss
        return 1
    elif team in expected_loss:
        return 0
    else:
        return None  # For the DenverBroncos/nan team
    
df['expected'] = df['subreddit'].apply(lambda x: assign_expected(x)) # creating new column 'expected' for the expected result
```

```python
# Spread from fanduel
spread = {'AZCardinals' : 7.0,'buffalobills' : -2.5,'panthers' : 3.5,'ChicagoBearsNFL' : -1.0,'bengals' : -2.5,'DenverBroncos' : -4.0,'Texans' : 10.0,'Colts' : 4.5,'KansasCityChiefs' : -6.5,'Chargers' : -2.5,'minnesotavikings' : -6.0,'Patriots' : 3.5,'NYGiants' : 3.5,'steelers' : 2.5,'Seahawks' : -4.5,'Tenesseetitans' : 3.0, 'falcons' : -3.5,'ravens' : -10.0,'Browns' : 2.5,'cowboys' : -3.5,'detroitlions' : 6.5,'GreenBayPackers' : 1.0,'Jaguars' : -4.5,'raiders' : 4.0,'LosAngelesRams' : 4.5,'miamidolphins' : 2.5,'Saints' : -3.0,'nyjets' : 2.5,'eagles' : -3.5,'49ers' : -2.5,'buccaneers' : 6.0,'Commanders' : -7.0}
print(len(spread))
```

```python
def assign_spread(team):     # function for assigning spread
    if team in spread:       # Simplyl assigns the spread from the dictionary
        return spread[team]
    else:
        return None # For the DenverBroncos/nan team
    
df['spead'] = df['subreddit'].apply(lambda x: assign_spread(x)) # Updates dataframe with a column 'spread' for the spread.
```
