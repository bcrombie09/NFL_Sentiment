```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv(r"initial_dataset.csv")
df2.head()
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.set_theme(style="whitegrid")
plt.xticks(fontsize = 8, rotation=60)
plt.title("Number of Comments per Subreddit")
comment_counts = df2['subreddit'].value_counts().reset_index()
comment_counts.columns = ['subreddit', 'count']

sns.barplot(x='subreddit', y='count', data=comment_counts, palette='viridis')
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.set_theme(style="whitegrid")
plt.xticks(fontsize = 8, rotation=60)
sns.barplot(x=df2['subreddit'], y=df2['compound'], palette="rocket").set(title="Average Compound Scores of Each Team")
```

```python
df2['comment_length'] = df2['comment'].astype(str).apply(lambda x: len(x))
sns.set_theme(style="ticks")
sns.pairplot(df2[['score','compound','comment_length']])
```
